# 1. login()

```python
def login():
    username=driver.find_element_by_name("username")
    username.send_keys("SAMPLE")    ## here replace sample by your user-name
    
    password=driver.find_element_by_name("password")
    password.send_keys("SAMPLE")   ### here replace sample by your password
    
    ### clicking on submit button
    
    button=driver.find_element_by_xpath("//*[@id='react-root']/section/main/article/div[2]/div[1]/div/form/div[4]/button")
    button.submit()
    
    time.sleep(6)
    
    # to handle notifications pop-up(may depend on individual id)
    
    try:
        driver.find_element_by_xpath("/html/body/div[4]/div/div/div[3]/button[2]").click() 
    except NoSuchElementException:
        pass

login()
```

# 2.search function for food

```python
def search_food():
    
    driver.get("https://www.instagram.com/")
    
    ## finding search bar and sending "food" as keys
    
    field=driver.find_element_by_xpath("//*[@id='react-root']/section/nav/div[2]/div/div/div[2]/input")
    field.send_keys("food")
    
    time.sleep(2)
    ig_handles=driver.find_elements_by_xpath("/html/body/div[1]/section/nav/div[2]/div/div/div[2]/div[2]/div[2]/div/a")
        
    print("INSTAGRAM HANDLES :")
    print()
    
    for i in ig_handles:
        
        ## removing hashtags and locations 
        ## printing only ig_handles
        
        if ("explore/tags" not in str(i.get_attribute("href"))) and "explore/locations" not in str(i.get_attribute("href")):
            instagram_handle=i.text.strip().split('\n')
            print(instagram_handle[0].strip())
    
    ## deleting what has been typed in search bar!
    
    to_clear=driver.find_element_by_xpath("//*[@id='react-root']/section/nav/div[2]/div/div/div[2]/input")
    to_clear.send_keys(Keys.CONTROL+"a")
    to_clear.send_keys(Keys.DELETE)
    
search_food()
```

# 5.1
## Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.

#### visit webdriver session to see how it is working after executing this script

```python
def like():
   
    
    ###SCROLLING to have 30 posts    
    
    current=driver.execute_script("return document.body.scrollHeight;")
    
    for i in range(2):
        driver.execute_script("window.scrollTo(0,arguments[0]);",current)
        time.sleep(8)
        new=driver.execute_script("return document.body.scrollHeight;")
        current=new
        
    
    time.sleep(2)
    
    ## saving posts in a list
    
    global posts     ## so that can be easily used while unliking
    
    posts=driver.find_elements_by_xpath('//*[@id="react-root"]/section/main/div/div[3]/article/div[1]/div//a')
    
    ## running through each post(TOP 30)

    for i in range(30):
        
        posts[i].click()
        
        time.sleep(4)
        
        like=driver.find_element_by_xpath('//span[contains(@class,"fr66n")]/button')
        svg=driver.find_element_by_xpath('//span[contains(@class,"fr66n")]/button/*[name()="svg"]')
    
        ## checking if already liked!
    
        if svg.get_attribute("aria-label")=="Unlike":
            print(i+1,"th post is already liked")
        else:
            like.click()
        
        ### finding close button
        
        close=driver.find_element_by_xpath("/html/body/div[4]/div[3]/button")
        close.click()
        time.sleep(1)
        
like()
```

# 5.2  
## Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
def unlike():
    for i in range(30):
        posts[i].click()
        time.sleep(1)
        unlike=driver.find_element_by_xpath('//span[contains(@class,"fr66n")]/button')
        svg=driver.find_element_by_xpath('//span[contains(@class,"fr66n")]/button/*[name()="svg"]')
    
        ## checking if already liked!
    
        if svg.get_attribute("aria-label")=="Like":
            print(i+1,"th post is already Unliked")
        else:
            unlike.click()
        
        ### finding close button
        
        close=driver.find_element_by_xpath("/html/body/div[4]/div[3]/button")
        close.click()
        time.sleep(1)

unlike()
```

# 6. Extract list of followers

## 6.1 Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.

```python
## for "So Delhi"

def so_delhi_followers():
    driver.get("https://www.instagram.com/")
    
    search=driver.find_element_by_xpath("//*[@id='react-root']/section/nav/div[2]/div/div/div[2]/input")
    
    search.send_keys("sodelhi")
    
    time.sleep(2)
    
    link=driver.find_element_by_xpath('//*[@id="react-root"]/section/nav/div[2]/div/div/div[2]/div[2]/div[2]/div/a')
    
    link.click()
    
    time.sleep(2)
    
    link=driver.find_element_by_partial_link_text("followers")
    
    link.click()
    
    time.sleep(2)
    
    while True:
        
        driver.find_element_by_class_name("isgrP").click() ## switching to followers list 
        
        driver.find_element_by_tag_name("body").send_keys(Keys.END) ### scrolling using End Key
        
        li=driver.find_elements_by_class_name("wo9IH")
        
        if len(li)>=500:
            break
    
    print("Followers of 'So Delhi'")
    
    ## printing usernames of followers 
    
    for i  in range(500):
        print((i+1),".",(li[i].text.strip()).split("\n")[0].strip())
      
    ## closing followers list
    
    driver.find_element_by_xpath('/html/body/div[4]/div/div[1]/div/div[2]/button').click()
    
so_delhi_followers()
    
```

```python
## for "foodtalkindia"

def ftf_followers():
    driver.get("https://www.instagram.com/")
    
    search=driver.find_element_by_xpath("//*[@id='react-root']/section/nav/div[2]/div/div/div[2]/input")
    
    search.send_keys("foodtalkindia")
    
    time.sleep(2)
    
    link=driver.find_element_by_xpath('//*[@id="react-root"]/section/nav/div[2]/div/div/div[2]/div[2]/div[2]/div/a')
    
    link.click()
    
    time.sleep(2)
    
    link=driver.find_element_by_partial_link_text("followers")
    
    link.click()
    
    time.sleep(2)
    
    while True:
        
        driver.find_element_by_class_name("isgrP").click() ## switching to followers list 
        
        driver.find_element_by_tag_name("body").send_keys(Keys.END) ### scrolling using End Key
        
        li=driver.find_elements_by_class_name("wo9IH")
        
        if len(li)>=500:
            break
    
    print("Followers of 'FoodTalkIndia'")
    
    global followers  ## making followers list globally so that it can be used in next question
    
    followers=[]
    
    ## printing usernames of followers 
    
    for i  in range(500):
        followers.append((li[i].text.strip()).split("\n"))
        print((i+1),".",(li[i].text.strip()).split("\n")[0].strip())
    
    ## closing followers list
    driver.find_element_by_xpath('/html/body/div[4]/div/div[1]/div/div[2]/button').click()
    

ftf_followers()
```

## 7. Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -

### 7.1 If You have already seen the story.
### 7.2 Or The user has no story.
### 7.3 Or View the story if not yet seen.

```python
def check_cn_story():
    driver.get("https://www.instagram.com/")
    time.sleep(3)
     
    list_stories=(driver.find_elements_by_xpath('//*[@id="react-root"]/section/main/section/div[1]/div[1]/div/div/div/div/ul/li'))[2:]
    list_canvas=driver.find_elements_by_xpath('//*[@id="react-root"]/section/main/section/div[1]/div[1]/div/div/div/div/ul/li//canvas')
    
    count=0
    for story in list_stories:
        
        if story.text=="coding.ninjas":
            if list_canvas[count].get_attribute("height")=="66":
                print("Visit Web driver Session, Story has been opened there!")
                story.click()
            else:
                print("Story has been Seen Already!")
            break
        count+=1
    else:
        print("Story Not Available!")

check_cn_story()
```
